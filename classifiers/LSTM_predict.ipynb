```python
max_length = 200

padding_type='post'
trunc_type='post'

with open('tokenizer.pickle', 'rb') as handle:
    trained_tokenizer = pickle.load(handle)


# TODO: REPLACE WITH ACTUAL TEXT IN FOMRAT WE WANT

file = open('../dataset/news_pickles/train_test_val_Human_posTags', 'rb')
X_human_train_posTags, X_human_test_posTags, X_human_val_posTags = pickle.load(file)
file.close()

file = open('../dataset/news_pickles/train_test_val_GPT_posTags', 'rb')
X_gpt_train_posTags, X_gpt_test_posTags, X_gpt_val_posTags = pickle.load(file)
file.close()

predict_text = np.hstack((X_human_train_posTags, X_gpt_train_posTags))

predict_sequences = trained_tokenizer.texts_to_sequences(predict_text)
test_padded = pad_sequences(predict_sequences, padding='post', maxlen=max_length)


predictions = model.predict(test_padded)
# Convert probabilities into binary outputs
binary_predictions = [1 if prob >= 0.5 else 0 for prob in predictions]
binary_predictions
```
